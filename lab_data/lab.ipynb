```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
file_path = "1_2006-2018 EPL stats.csv"

df = pd.read_csv(file_path)
```

```python
df.info()

df.describe()
```

```python
filtered_df = df[df['goals'] > 50]
filtered_df.head()
```

```python
sorted_df = df.sort_values(by='wins', ascending=False)
sorted_df.head()
```

```python
grouped = df.groupby('season')['goals'].mean()
grouped
```

```python
def filter_by_range(dataframe, column, min_value, max_value):
    """Фільтрує DataFrame за діапазоном значень у вказаній колонці"""
    return dataframe[(dataframe[column] >= min_value) & (dataframe[column] <= max_value)]

filtered_custom = filter_by_range(df, 'losses', 10, 100)
filtered_custom.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['goals'], kde=True, bins=30)
plt.title("Розподіл значень у колонці 'goals'")
plt.show()
```

```python
pivot_table = df.pivot_table(values='wins', index='team', aggfunc='sum')
pivot_table
```

```python
filtered_wins = df[df['wins'] > 20]
print(f"Кількість команд з більше ніж 20 перемог: {len(filtered_wins)}")
filtered_wins.head()
```

```python
sorted_goals = df.sort_values(by='goals', ascending=False)
sorted_goals.head()
```

```python
grouped_season = df.groupby('season')['goals'].mean()
grouped_season.plot(kind='bar', figsize=(12, 6), color='skyblue', title="Середня кількість голів за сезон")
plt.ylabel("Середня кількість голів")
plt.xlabel("Сезон")
plt.show()
```

```python
pivot_table = df.pivot_table(values=['wins', 'losses'], index='team', aggfunc='sum').sort_values(by='wins', ascending=False)
pivot_table.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total_yel_card'], kde=True, bins=20, color='orange')
plt.title("Розподіл жовтих карток")
plt.xlabel("Кількість жовтих карток")
plt.ylabel("Частота")
plt.show()
```
